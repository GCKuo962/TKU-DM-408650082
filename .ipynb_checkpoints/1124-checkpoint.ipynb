```python
df1 = pd.read_csv('score_1.csv')
```

```python
df2 = pd.read_csv('score_3.csv')
```

```python
df1.append(df2, ignore_index=True)
```

## HW1: 請用上週教的方法將NaN補為60,再次印出表格

```python
hw1 = df1.append(df2, ignore_index=True)
hw1.fillna(60, inplace = True)
hw1
```

## 資料融合

```python
df1 = pd.read_csv('score_1.csv')
```

```python
df2 = pd.read_csv('score_4.csv')
```

```python
pd.merge(df1,df2)
```

```python
df1 = pd.read_csv('score_1.csv')
```

```python
df2 = pd.read_csv('score_4.csv')
```

```python
pd.merge(df1,df2, how='outer')
```

## HW2: 請用上週教的方法將NaN補為該科平均值,再次印出表格

```python
hw2 = pd.merge(df1,df2, how='outer')
hw2['國文'] = hw2['國文'].fillna(hw2['國文'].mean())
hw2['數學'] = hw2['數學'].fillna(hw2['數學'].mean())
hw2['自然'] = hw2['自然'].fillna(hw2['自然'].mean())
hw2
```
