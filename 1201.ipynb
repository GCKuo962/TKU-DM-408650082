# 糖尿病 / BMI預測

```python
import matplotlib.pyplot as plt            # 繪製函式庫
import numpy as np                         # 矩陣函式庫
from sklearn import datasets, linear_model # 線性迴歸函式庫
```

```python
# 取得糖尿病的數據
diabetes = datasets.load_diabetes()

# 只取第三個特徴值BMI
diabetes_X = diabetes.data[:, np.newaxis, 2]

# 切分特徴值BMI
diabetes_X_train = diabetes_X[:-20]        # 切割 0 到最後 20 筆特微給訓練用
diabetes_X_test = diabetes_X[-20:]         # 切割最後 20 筆特微給訓練用

# 切分答案
diabetes_y_train = diabetes.target[:-20]   # 切割 0 到最後 20 筆答察給訓練用
diabetes_y_test = diabetes.target[-20:]    # 切割最後 20 筆答案給訓練用
```

```python
# 計算
regr = linear_model.LinearRegression()        # 建立線性迴歸
regr.fit(diabetes_X_train, diabetes_y_train)  # 訓練

# 係數
print("Coefficients: \n", regr.coef_)

# 均方誤差
print("Mean squared error: %.2f"
      % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))

# 顯示方差分數:1 是完美預測
print("Variance score: %.2f" % regr.score(diabetes_X_test, diabetes_y_test))

# 繪圖
plt.scatter(diabetes_X_test, diabetes_y_test, color = "black") # 畫出測試黑點
plt.plot(diabetes_X_test, regr.predict(diabetes_X_test), color = "blue",
         linewidth = 3) # 畫出測試預測的線性迴歸答案

plt.xticks(())
plt.yticks(())

plt.show() # 顯示
```

```python
import numpy as np

from sklearn.model_selection import train_test_split # 匯入 K-means 函式庫
from sklearn.cluster import KMeans
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics

X = np.array([[1,1], [1.1,1.1], [1.2,1.2],
             [2,2], [2.1,2.1], [2.2,2.2]])

y = [1,1,1,
     0,0,0]

kmeans = KMeans(n_clusters = 2, random_state = 0).fit(X) # KMeans演算法 指定 2 類資料

print("集群中心的坐標:", kmeans.cluster_centers_)                    # 取得集群中心的座標
print("預測:", kmeans.predict(X))                                   # 預測
print("實際:", y)                                                   # 真實答案
print("預測[1,1],[2.3,2.1]:", kmeans.predict([[1,1], [2.3,2.11]]))  # 預測資料

plt.axis([0, 3, 0, 3])              # 用圖片顯示
plt.plot(X[:3, 0], X[:3, 1], "yx")  # 黃色點
plt.plot(X[3:, 0], X[3:, 1], "g.")  # 綠色點
plt.plot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], "ro") # 顯示中心點

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics

# 取得鳶尾花的資料
iris = datasets.load_iris()

# 拆分資料 80:20，訓練與測試用 2 類標籤答案
iris_X_train, iris_X_test, iris_y_train, iris_y_test = train_test_split(iris.data, iris.target, test_size = 0.2)

# KMeans演算法 2 類標籤答案
kmeans = KMeans(n_clusters = 2)
kmeans_fit = kmeans.fit(iris_X_train) # 訓練，沒有標籤答案

print("實際", iris_y_train)       # 顯示標筹案
print("預測", kmeans_fit.labels_) # 顯示預測的答案

# 調整標籤的數字
iris_y_train[iris_y_train == 1] = 11
iris_y_train[iris_y_train == 0] = 1
iris_y_train[iris_y_train == 11] = 0
print("調整", iris_y_train)       # 顯示調整後的預測

score = metrics.accuracy_score(iris_y_train, kmeans.predict(iris_X_train))
print("準確率:(0:f)".format(score)) # 顯示準確率
```

# HW1 K - Means應用：信用卡客戶分群

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace = True)
df
```

```python
km = KMeans(n_clusters = 3)
km.fit(df)
km.labels_
```

```python
df['類別'] = km.labels_
df.head()
```

```python
df2 = df[df['類別'] == 0]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別'] == 1]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別'] == 2]
df3 = df2.iloc[0:30, :]
df3
```

```python
for n in range(2, 15):
  km = KMeans(n_clusters = n)
  km.fit(df)
  metric = calinski_harabasz_score(df, km.labels_)
  print('群組數量：{}，評分：{}'.format(n, metric))
```

# HW2 DBSCAN應用：信用卡客戶分群

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace = True)
df
```

```python
scaler = StandardScaler()
scaler.fit(df)
dfScaled = scaler.transform(df)
dbscan = DBSCAN(eps = 0.82, min_samples = 5)
dbs = dbscan.fit(dfScaled)
dbs.labels_
```

```python
df['類別'] = dbs.labels_
df.head()
```

```python
df2 = df[df['類別'] == 0]
df3 = df2.iloc[0:30, :]
df3
```
